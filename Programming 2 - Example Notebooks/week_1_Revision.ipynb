## Conditions and Loops

### `if-elif-else` Conditions

Conditional statements `if`-`elif`-`else` can be used to control branching in your code. They determine which block of code is executed based on whether a condition is `True` or `False`. Check out the flowchart below 👇, and note that, only the first `True` condition will execute, and the rest are skipped.

<p align="center">
<img src="http://binghuan.li/Notes/files/BIOE50010/outgoing/if_else_cond.svg" alt="if_else" width="800" height="auto">
</p>


```python
## Programme 3-6: grader.py, from *Starting Out With Python (Global Ed.)*

## This program gets a numeric test score from the
## user and displays the corresponding letter grade.

# Named constants to represent the grade thresholds
A_SCORE = 90
B_SCORE = 80
C_SCORE = 70
D_SCORE = 60

# Get a test score from the user.
score = int(input('Enter your test score: '))

# Determine the grade.
if score >= A_SCORE:
    print('Your grade is A.')
else:
    if score >= B_SCORE:
        print('Your grade is B.')
    else:
        if score >= C_SCORE:
            print('Your grade is C.')
        else:
            if score >= D_SCORE:
                print('Your grade is D.')
            else:
                print('Your grade is F.')
```

### `while` Conditions

A `while` condition in Python is used to repeatedly execute a block of code as long as the condition is `True`, as shown by the flow chart below 👇. Be really careful if the condition never becomes `False`, the loop will run forever (infinite loop), and your program may hang or crash!🤦

<p align="center">
<img src="http://binghuan.li/Notes/files/BIOE50010/outgoing/while_cond.svg" alt="while" width="400" height="auto">
</p>

```python
## Programme 4-2: grader.py, from *Starting Out With Python (Global Ed.)*

# This program assists a technician in the process
# of checking a substance's temperature.

# Named constant to represent the maximum temperature.
MAX_TEMP = 102.5

# Get the substance's temperature.
temperature = float(input("Enter the substance's Celsius temperature: "))

# As long as necessary, instruct the user to adjust the thermostat.
while temperature > MAX_TEMP:
    print('The temperature is too high.')
    print('Turn the thermostat down and wait')
    print('5 minutes. Then take the temperature')
    print('again and enter it.')
    temperature = float(input('Enter the new Celsius temperature: '))

# Remind the user to check the temperature again in 15 minutes.
print('The temperature is acceptable.')
print('Check it again in 15 minutes.')
```

### `for` Loops

A `for` loop repeats a block of code for each item in an iterable sequence, for example, a `list`, a `str`, or [`range()`](https://docs.python.org/2/library/functions.html#range). Since the number of iterations is usually known in advance, you generally do not need to worry about infinite loops. However, you cannot iterate over a *non-iterable* data in Python, for example, an `int` (which makes no sense doesn't it? 🙅)

<p align="center">
<img src="http://binghuan.li/Notes/files/BIOE50010/outgoing/for_loop.svg" alt="for" width="600" height="auto">
</p>

```python
#Example: Calculate the average of a list of numbers

scores = [85, 92, 78, 95, 88, 70, 90]
total_score = 0

for score in scores:
  total_score += score

average_score = total_score / len(scores)

print(f"The list of scores: {scores}")
print(f"The total score: {total_score}")
print(f"The average score: {average_score:.2f}")
```

## Functions and Modularisation

The true elegance behind the functions are closely linked to the idea of **modules**. In modular programming, functions are put together to make
up one executable program.

- Each function is separately defined → making it reusable ♻️

- Functions are triggered serially in a main script ▶️


```python
## Modified from Programme 4-2: grader.py, from *Starting Out With Python (Global Ed.)*

# This program allows the user to choose various
# geometry calculations from a menu.

import math

# The area function accepts a circle's radius as an
# argument and returns the area of the circle.
def circle_area(radius):
    return math.pi * radius**2

# The circumference function accepts a circle's
# radius and returns the circle's circumference.
def circle_circumference(radius):
    return 2 * math.pi * radius

# The area function accepts a rectangle's width and
# length as arguments and returns the rectangle's area.
def rectangle_area(width, length):
	return width * length

# The perimeter function accepts a rectangle's width
# and length as arguments and returns the rectangle's
# perimeter.
def rectangle_perimeter(width, length):
	return 2 * (width + length)

# The display_menu function displays a menu.
def display_menu():
    print(' MENU')
    print('1) Area of a circle')
    print('2) Circumference of a circle')
    print('3) Area of a rectangle')
    print('4) Perimeter of a rectangle')
    print('5) Quit')

# The main function.
def main():

  # Constants for the menu choices
  AREA_CIRCLE_CHOICE = 1
  CIRCUMFERENCE_CHOICE = 2
  AREA_RECTANGLE_CHOICE = 3
  PERIMETER_RECTANGLE_CHOICE = 4
  QUIT_CHOICE = 5

  # The choice variable controls the loop and holds the user's menu choice.
  choice = 0

  while choice != QUIT_CHOICE:
      # display the menu.
      display_menu()

      # Get the user's choice.
      choice = int(input('Enter your choice: '))

      # Perform the selected action.
      if choice == AREA_CIRCLE_CHOICE:
        radius = float(input("Enter the circle's radius: "))
        print('The area is', circle_area(radius))

      elif choice == CIRCUMFERENCE_CHOICE:
        radius = float(input("Enter the circle's radius: "))
        print('The circumference is', circle_circumference(radius))

      elif choice == AREA_RECTANGLE_CHOICE:
        width = float(input("Enter the rectangle's width: "))
        length = float(input("Enter the rectangle's length: "))
        print('The area is', rectangle_area(width, length))

      elif choice == PERIMETER_RECTANGLE_CHOICE:
        width = float(input("Enter the rectangle's width: "))
        length = float(input("Enter the rectangle's length: "))
        print('The perimeter is', rectangle_perimeter(width, length))

      elif choice == QUIT_CHOICE:
        print('Exiting the program. . .')

      else:
        print('Error: invalid selection.')

# Call the main function.
main()
```
