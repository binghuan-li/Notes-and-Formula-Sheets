## File I/O 🗄

File input and output is a basic capatability that the Python interpreter holds to read from and write to files on your computer.

In the slides, we have seen the logic flow to handle a file in Python, to summarise:

1. Open the file using `open()`, you need to supply the file name (and path), also the mode (read or write).

2. If your choose the **read** mode, you can now read from the file: `read()`, `readline()`, `readlines()`.

3. ...Or, if your choose the **write** mode, you can write to the output stream using `write()`, `writeline()`, `writelines()`.

4. Never forget to **close** the I/O stream when you no longer need it!

The first two examples below shall demonstate the usage of `write()` and `writelines()`.


````python
# This program saves a list of strings to a file.

def main():
    # Create a list of strings.
    cities = ['New York', 'Boston', 'Atlanta', 'Dallas']

    # Open a file for writing. Modify the path if need.
    outfile = open('cities.txt', 'w') # <= mode 'w' for writing

    # Write the list to the file.
    for item in cities:
        outfile.write(item + '\n') # <== what does '\n' do?

    # Close the file.
    outfile.close()

# Call the main function.
main()
````

❓ **What does `\n` do?** The character `\n` appended to a string is known as the *escape sequence* representing a new line. `\n` is not printed as a text, but instead it is intrepreted as "starting a new line after the current string".

The key here is the backslash `\` - in an output stream, Python treat it as a special formatting instruction. One more example is `\t`, which introduces a tab (4 character spaces)!

---

We have seen the file output, what about **reading** those city names were wrote to the file back to our Python programme? Huh, we shall now see how to read the file contents and process them so that we can use these data again!

````python
# This program reads a file's contents into a list.

def main():
    # Open a file for reading. Modify the path if need.
    infile = open('cities.txt', 'r')

    # Read the contents of the file into a list.
    cities = infile.readlines()

    # Close the file.
    infile.close()

    # Strip the \n from each element.
    index = 0
    while index < len(cities):
        cities[index] = cities[index].rstrip('\n')
        index += 1

    # Print the contents of the list.
    print(cities)

# Call the main function.
main()
````

As you have seen, reading data from a file is often accompanied by some kind of processing - such as stripping unwanted characters, and format them, so they can be used later in the programme.

## String Methods

We are now going to talk about the string methods.

❓Wait... "what is a *method*?" I hear you ask. Well, a method refers to a function, but more than a function, **a method is a function associated with a specific data type**. For example, string methods are the functions that can be only applied to a `str` type variable, but not others.

For the syntax, a method comes with a dot (`.`). For example, `.split()`, `.strip()`. When calling, it is applied to `str` type variable, such as `text.split()`, `text.strip()`.

---
👍So... yes! String methods are the functions to perform common string operations such as *splitting*, *trimming*, *searching*, or *replacing*. You may find these methods particularly useful when *e.g.* clean or transform text data effectively.

Below is a table of frequently used string methods along with their descriptions:

| **Method**               | **Description** |
|--------------------------|-----------------|
| `startswith(substring)`  | The method returns **true** if the string starts with *substring*. |
| `endswith(substring)`    | The method returns **true** if the string ends with *substring*. |
| `find(substring)`        | The method returns the lowest index in the string where *substring* is found. If *substring* is not found, the method returns -1. |
| `replace(old, new)`      | The method returns the string with all instances of *old* replaced by *new*. |
| `lstrip(char)`           | The method returns a copy of the string with the specified character (*char*) that appear at the beginning (**left**) of the string removed. |
| `rstrip(char)`           | The method returns a copy of the string with the specified character (*char*) that appear at the end (**right**) of the string removed. |
| `split(delimiter)`       | The method returns a list containing the words in the string separated by the specified *delimiter*, by default the *delimiter* is a whitespace. |


````python
## Example

# Original text with leading and trailing spaces, a comma, and an asterisk
text = "  The quick brown fox, jumps over * the lazy dog.  "

# first strip out the spaces
text_1 = text.strip()
print(text_1)

# split by comma
text_2 = text_1.split(',')
print(text_2)

# split the second part (index 1) by asterisk
text_3 = text_2[1].split('*')
print(text_3)
````

## List Methods

List methods are similar to the string methods. 🤓 Let me bet, you may have been fiarly famililar with some - for example, `append()` is **extremely** useful for adding a new item to the end of an existing list - you will like it!

Below is a table of frequently used list methods along with their descriptions:

| **Method**                | **Description** |
|---------------------------|-----------------|
| `append(item)`            | Adds *item* to the end of the list. |
| `index(item)`             | Returns the index of the first element whose value is equal to *item*. A `ValueError` exception is raised if *item* is not found in the list. |
| `insert(index, item)`     | Inserts *item* into the list at the specified *index*. |
| `sort()`                  | Sorts the items in the list so they appear in ascending order (from the lowest value to the highest value). |
| `remove(item)`            | Removes the first occurrence of *item* from the list. A `ValueError` exception is raised if *item* is not found in the list. |
| `reverse()`               | Reverses the order of the items in the list. |


````python
## Example

# Initialize a list
fruits = ['apple', 'banana', 'cherry']

# append(item): Add an item to the end
fruits.append('date')
print("After append:", fruits)

# index(item): Find the index of an item
banana_index = fruits.index('banana')
print("Index of 'banana':", banana_index)

# insert(index, item): Insert an item at a specific position
fruits.insert(2, 'blueberry')
print("After insert:", fruits)

# sort(): Sort the list alphabetically
fruits.sort()
print("After sort:", fruits)

# remove(item): Remove a specific item
fruits.remove('cherry')
print("After remove:", fruits)

# reverse(): Reverse the order of the list
fruits.reverse()
print("After reverse:", fruits)
````
