## Positional Argument *v.s.* Keyword Argument 🆚

You probably already know that a **function argument** refers to the **input** of that function.

For example, imagine you are going to retrieve student info from a database through a Python function, `get_student_info(name, cid)`; here, `name` and `cid` are the arguments passed into the function, allowing the function to do something useful - like fetch a student record.

Easy. When I call the function... I can either do
```
get_student_info('Webster', 1234567)
```
or...
```
get_student_info(name='Webster', cid=1234567)
```

❓Huh... "What are the difference between these two function callers?" I hear you ask. Well, in the first version, the values are matched to argument based on their **position** in the function definition — `Webster` goes to `name`, and `1234567` to `cid`, in that order. Whereas in the second version, you're telling Python exactly which value goes with which argument by using the names (**keyword**).

That's it! 🙂





## Non-Keyword Argument Functions

Well... here comes a *slightly tricky* need. Suppose Mr. H (you-know-who)  shows up one day (on behalf of our Student Office), and kindly asks "can you you to revise the `get_student_info` function, so that it can handle the case where multiple student names can be passes into the function and searched in one go?"🚶

"Wait, wait...", you replied, "did you just ask me to tweak the function to accept a flexible set of inputs?" 🤔

"Indeed."

That is where **non-keyword argument functions** comes in!

The non-keyword argument functions (or simply `*args` for short), allows us to pass a variable number of non-keyword, positional arguments to a function — like a flexible input list that you don't have to predefine. It simply takes the format:

```Python
def my_function_name(*arg):
  ...
```

---
**Comments**:
- The asterisk `*` is known as the unpacking operator.
- All arguments (input) are collected and packed into `arg`, which is of the type `tuple`. To retrieve each of the argument, you need a `for` loop to iterate through `arg`.
- If any positional arguments present, they must come before `*args`. In other words, `function_name(positional_args, *args)`.
---

So, here is your solution:

````python
def get_student_info(*cid):
  print(f'Woww what a heave job... We have {len(cid)} students to search.')
  for id in cid:
    print(f'Now searching for cid: {id}... Done!')

# Call the function.
get_student_info(1234567, 2345678, 3456789)
````

## Keyword Argument Functions

"Phew... Almost done", You think.

But then, the next thought hits you:
>"Ahh! To make it really user-friendly, I need to let users submit as many `name-ID` pairs as they want — without forcing them into a rigid format."

Isn't there a **keyword argument** feature that can take as many input pairs as possible? 🤔 And, we do not even to change our code to much!

For `**kwarg`, the function definition format reads

```Python
def my_function_name(**kwarg):
  ...
```

---
**Comments**:
- All `*kwargs` are collected and packed into a dictionary (`dict`) and stored as `key:value` pairs. Hence, to retrieve each of the argument pair, we use a `for` loop to iterate through the dictionary (how?).

- for example, if I want both the *key* and *value* in a pair, simply use the `item()` function. Check out this [useful reference](https://www.w3schools.com/python/python_dictionaries_loop.asp)!

- Positional arguments and `*args` must come before `**kwargs`.
---

````python
def get_student_info(**cid):
  print(f'Woww what a heave job... We have {len(cid)} students to search.')
  for name, id in cid.items(): # use .item() to loop over a dictionary
    print(f'We are searching for {name}: {id}... Done!')

# Call the function.
get_student_info(Hermione=7654321, Harry=1122334, Ron=9988776)
````
