1. a (custom-made) data type - just like built-in types such as `int`, `str`, `list`.

## Box 📤

As we kicked off this new term, you received a welcome pack from our very own [Bioengineering Society](https://www.instagram.com/icl_bgsoc/?hl=en). 🆒 It came in a little neat box - but what's inside? 🤔

In this example, we'll use that box to explore how objects can hold both data (like the box's color or the items inside) and behavior (like opening or closing it).

```python
# Define the Box class
class Box:

  # method 1: initialise the Box object ("constructor")
  def __init__(self, color, items):
      self.color = color
      self.items = items
      self.is_open = False # <--- the default state


  # method 2: run description of the box
  def describe_box(self):
      print(f"Aha, this is a {self.color} box.")
      print(f"Let me see what's inside...")

      if not self.is_open:  # if the box is closed
        print("Nahhh... The box is closed. Open it first!\n")
      else:                 # if the box is opened
        for item in self.items:
            print(f"Ohh! I got a {item}!")
        print()


  # method 3: open the box
  def open_box(self):
      if not self.is_open:  # if the box is closed, open it
          self.is_open = True
          print("The box is now open.\n")
      else:                 # if the box is opened, do not open it twice
          print("The box is already open.\n")


  # method 4: close the box
  def close_box(self):
    if self.is_open:        # if the box is opened, close it
      self.is_open = False
      print("The box is now closed.\n")
    else:                   # if the box is closed, do not close it twice
      print("The box is already closed.\n")
```

```python
# ------------driver code to use the class Box------------

# Instantiate the obj.
blue_box = Box(color="blue",
               items=['Splice magazine', 'boat party invitation', 'selfie of Mr. H'])

# Print the description of the box object (at the closed state)
blue_box.describe_box()

# Open the box
blue_box.open_box()
blue_box.open_box()

# Print the description of the box object (at the opened state)
blue_box.describe_box()

# Now, close the box
blue_box.close_box()
blue_box.close_box()
```

---

In our example, we've already seen two key OOP ideas in action!

- With **abstraction**: When you use the box, you donit worry about how it opens or lists its contents - you just call methods like `open_box()` or `describe_box()`. The complex logic stays hidden inside the class definition.

- With **encapsulation**: The box keeps its data (`color`, `items`, `is_open`) bundled safely inside. We can only interact with it through methods.

Together, these ideas help us build programs that are tidy, easy to use, and hard to break - just like a well-designed gift box! 🎁💡

## Operator Overloading

👓 **A tiny discovery**: if you try to print the object...

```python
# Print the object's memory address
print(f'The address of the box object is: {blue_box}')
```

Surprise?! You probably see a bunch of random-looking text such as `<__main__.Box object at 0x79f6206fa540>` - this is just Python showing you the memory address of the `blue_box` object (where it "lives" in your computer's memory).

---

🤔 "Huh... this is not a helpful message." I hear you murmuring.

To make your object say something more meaningful, like its *color*, *state*, or even a short description, you have two options:

1. Write your own method (like `describe_box()`) to print out what you want; or

2. Give your class a little magic power 🔥 **by defining a special (magic) method `__str__()`**: This tells Python how your object should introduce itself when you do `print(blue_box)`.

Let's sprinkle in some magic and update our class definition! ✨

```python
# Modifies class definition of Box
class Box:

  # method 1: initialise the Box object ("constructor")
  def __init__(self, color, items):
      self.color = color
      self.items = items
      self.is_open = False # <--- the default state


  # method 2: run description of the box
  def describe_box(self):
      print(f"Let me see what's inside...")

      if not self.is_open:  # if the box is closed
        print("Nahhh... The box is closed. Open it first!\n")
      else:                 # if the box is opened
        for item in self.items:
            print(f"Ohh! I got a {item}!")
        print()


  # method 3: open the box
  def open_box(self):
      if not self.is_open:  # if the box is closed, open it
          self.is_open = True
          print("The box is now open.\n")
      else:                 # if the box is opened, do not open it twice
          print("The box is already open.\n")


  # method 4: close the box
  def close_box(self):
    if self.is_open:        # if the box is opened, close it
      self.is_open = False
      print("The box is now closed.\n")
    else:                   # if the box is closed, do not close it twice
      print("The box is already closed.\n")


  # method 5: __str__ that defines the message to print
  def __str__(self):
      return f"Aha, this is a {self.color} box."


# ---------------------------------------
# Instantiate the obj.
blue_box = Box(color="blue",
               items=['Splice magazine', 'boat party invitation', 'selfie of Mr. H'])

# Print the customised description of the box (rather than the memory address)
print(blue_box)
```

**Why is this?** ❓

As we briefly mentioned, `__str__` is one of Python's *special methods* - and it's special for a reason!

1. the name `__str__` has been *preserved* for a given purpose; You can't just redefine it to do something unrelated - Python expects it to describe how your object should look as text.

2. The goal of this method is to **return a string** that represents your object - it's what gets shown when you use `print(blue_box)`.

3. If you don't define `__str__`, Python falls back to its default behavior: showing the object's memory address.


Overall, defining special methods like `__str__` is called **operator overloading** - giving your objects custom behavior, or overriding the defaults Python would normally use.

---

**`__str__` isn't alone in this magical world** ✨ - Python has *many* special methods, all starting and ending with double underscores (`__`). For example:

1. Possibly, the most widely used magic method is `__init__`: you never call it, but it runs automatically when you create an object;

2. `__eq__` → defines equality (`==`);

3. `__add__` → defines addition (`+`).
