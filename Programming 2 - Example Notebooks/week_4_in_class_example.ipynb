Tired of messy number? This light exercise shows you how to line up numbers nicely using `f`-string formatting with `print()`.🌞 In this example, we right-align numbers with different floating points using 10 spaces, and the output looks just like a mini-table!

```python
numbers = [[1.28, 0.243, 2.326], [0.8624, 5, 4.6]]

# without using alignment
for i in range(len(numbers)):
  for j in range(len(numbers[i])):
    print(numbers[i][j], end='')
  print()

print('------------')

# Using alignment (right aligned, 10 spaces)
for i in range(len(numbers)):
  for j in range(len(numbers[i])):
    print(f'{numbers[i][j]:>10}', end='')
  print()
```
