## Mammals, Dogs, and the Magic of **Inheritance** 🐕

Imagine a grand kingdom of animals. At the top sits the wise ancestor class: Mammal, keeper of warm blood and mysterious mammalian noises. Every furry descendant **inherits** these characteristics, while deriving **unique** talents that set them apart from their mammalian siblings...

```python
# Superclass
class Mammal:
    def __init__(self, name):
        self.name = name

    def is_warm_blood(self):  # Common characteristic: warm-blooded
        return f"{self.name} is warm-blooded because science said so."

    def speak(self):            # Default sound for a generic mammal
        return "Grrrr! (some default sound for a generic mammal?!)"


# Subclass Dog inheriting from parent class Mammal
class Dog(Mammal):
    def __init__(self, name, color, favorite_toy):
        super().__init__(name)  # Initialize the superclass (Mammal)
        self.color = color
        self.favorite_toy = favorite_toy

    def speak(self):            # Overriding the speak method (polymorphism)
        return "Woof woof! (Pay attention to me!)"

    def fetch(self):
        return f"{self.name} zooms off to get the {self.favorite_toy}!"

    def description(self):
        return f"{self.name} is a {self.color} {self.__class__.__name__} who loves a good {self.favorite_toy}."


# ------------driver code to use the classes------------
# Instantiate the objects
Buddy = Dog("Buddy", "golden", "tennis ball")
print(Buddy.description())
print(Buddy.speak())
print(Buddy.fetch())
print(Buddy.is_warm_blood())
```

As you have seen above, class `Dog` *inherits* the attributes (variables) and functions (methods) which have been previsouly defined in its parent class `Mammal`. In this case, even the method `is_warm_blood()` is not defined in `Dog`, this method can be used in any objects belong to the `Dog` class.

However, both `Mammal` and `Dog` have implemented the method `speak()`. Even though the method name is the same in both classes, the version inside `Dog` overrides the one in `Mammal`. (In other words, when `speak()` is called on a `Dog` object, Python will automatically use the method defined in the `Dog` class, not the one from `Mammal`.) This behaviour is known as the *polymorphism*. ✨
