## **Variable Names, Objects, `copy`**

❓**Have you ever wondered what actually happens inside Python when you assign a value to a variable?**

When you create a variable in Python (*e.g.*, `a = [1, 2, 3]`), what *really* happens is that variable name (*e.g.*, `a`) is bound to objects (*e.g.*, `[1, 2, 3]`) in memory.

Interesting... 🤔 So, variables don't store data directly, they hold references to objects in memory (*a.k.a* "pointers"). When you access a variable by its name, Python retrieves (dereferences) the actual object through that memory reference.

"How can we believe your words? Isn't this too wild?" I hear you murmuring. Let us explore this through this quick example (using lists, for now)...



```python
# A list with nested structure
list_a = [[1, 2, 3], [4, 5, 6]]

# copy list_a to list_b
list_b = list_a

print('before modifying list_a...')
print("list_a:", list_a)
print("list_b:", list_b)

# Modify the nested list inside list_a
list_a[0][0] = 'X'

print('\nafter modifying list_a...')
print("list_a:", list_a)
print("list_b:", list_b)
```

Huh... 👀 In the code above, it looks like I just made a copy of `list_a` and assigned to a new variable name `list_b`. But why is it that when I modified an element in `list_a`, the change also appeared in `list_b`?!

Or... Python didn't even make a new list `list_b` when I do `list_b = list_a`? 😥

---

Indeed, this example deciphers how Python handles the variable names: instead of creating a fresh copy, both names, `list_a` and `list_b`,  are now **aliases** that refer to the exact same object in memory.

That's why, when you modify something inside `list_a` (like changing the '1' to 'X'), the change appears in `list_b` too! 💯

We refer to this type of copy as the **shallow copy** (or, it's simply an assignment).

### **Deep copy**

Having this said, if you actually want a new `list_b` that is 💯 independent from `list_a`, the function `deepcopy()` is surely the thing you are looking for.

A deep copy recursively duplicates every object inside the nested data structure, ensuring that changes in one have absolutely no effect on the other. 🆒 Let us see this example:

```python
import copy

# Original list with nested structure
list_a = [[1, 2, 3], [4, 5, 6]]

# copy list_a to list_b
list_b = copy.deepcopy(list_a)

print('before modifying list_a...')
print("list_a:", list_a)
print("list_b:", list_b)

# Modify the nested list inside the original
list_a[0][0] = 'X'

print('\nafter modifying list_a...')
print("list_a:", list_a)
print("list_b:", list_b)
```

###❗**Copying a Number**

Having said, when I copied a list, changing one affected the other. Is this behaviour the same for integers❓ Let us check...

```python
# Declare int_a and assign int_a to int_b
int_a = 12
int_b = int_a
print('before modifying int_a...')
print("int_a:", int_a)
print("int_b:", int_b)

# Modify int_a
int_a = 100
print('\nfter modifying int_a...')
print("int_a:", int_a)
print("int_b:", int_b)
```

What⁉ Isn't this a trick? 😟 Well... here is the key:

- In our earlier example, we used a `list`, which is a **mutable** object.

- This time, though, we're using an `int`, which is an **immutable** object.

This means, when reassigning a new value to `int_a`, the previous value `int_a` refers to (`12`, in our case) cannot be changed in place (because it is immutable) - you just can't "modify" the integer `12`. Instead, what really happened is Python *rebinds* the variable `int_a` to a new `int` object `100`.

But now, `int_b` is still happily pointing to the old object (`12`). 🙂

So... that's why changing an integer variable doesn't affect its “copy”. You didn't change the object itself - you just made the variable name point to a new one!

## **Handling Exceptions**

The `try`-`except`-`finally` cluase allows Python to catch and handle exceptions gracefully, ensuring that programs can respond to unexpected errors without crashing. 💪

- The `try` block contains the codes that will execute first, however, they may raise an exception.

- The `except` block defines how to handle specific types of exceptions that occur in the try block. You can specify one or more exception types or use a generic `except` (without specifying the error type) to catch all exceptions.

- The `finally` block contains code that will always execute, whether or not an exception occurs.

```python
counter = 0
usr_input_list = []

while counter < 5:

  try:
      # code that might cause an error
      usr_input = input("Enter a number: ")
      result = 10 / int(usr_input)
      print("Result:", result)

  except ValueError:
      # if converting input to int fails, e.g., user enters 'hello'
      print("Oops! That was not a valid number.")

  except ZeroDivisionError:
      # if the user enters 0 (division by zero)
      print("You can't divide by zero!")

  finally:
      # regardless of any exceptions raised, record the user input
      usr_input_list.append(usr_input)
      counter += 1

print("\nuser input list:", usr_input_list)
```

## (optional topic) **Composition in OOP**

> First things first, kindly note that this is a *relatively* advanced topic in OOP that we chose to omit in the main lectures for clarity and simplicity. The following information is provided for the sake of completeness, allowing those who wish to delve deeper into the subject to gain a comprehensive understanding. These are surely *not* examinable, but yet you will find them useful in the real coding practices.

The `Box` class contains an `item` attribute that can hold an object of `ToyTrain`. This code snippets demonstrates composition, where `Box` can interact with the `ToyTrain` object's methods, as if there is a toy train packed in a delivery box.

<p align="center">
<img src="http://binghuan.li/Notes/files/BIOE50010/outgoing/composition.png" alt="Composition" width="400" height="auto" class="center">
</p>


```python
# the ToyTrain class
class ToyTrain:
    def __init__(self, color:str, size:str, max_speed:float):
        self.color = color
        self.size = size
        self.max_speed = max_speed

    def print_descrption(self):
        print(f'>> Oh! I get a {self.size} {self.color} toy train! \n')

    def accelerate_train(self):
        print('>> The toy train is accelerating...')
        for i in range (10):
            print(f'instantaneous velocity now: {i*(self.max_speed/10):.2f} m/s...')
        print(f'The toy train has reached its maximum speed at {self.max_speed} m/s \n')


# the Box class
class Box():
    def __init__(self):
        self.is_open = False

    def open_the_box(self):
        if not self.is_open:
            self.is_open = True
            print('>> The box is now opened! \n')
        else:
            print('>> The box is already opened, no need to open again \n')

    def unpack_item(self, item:ToyTrain): # <= note that this type hinting
                                          # enforces `item` is an object of
                                          # the class ToyTrain!
        if not self.is_open:
            print('>> Open the box first! \n')
        else:
            # composition: the attribute 'item' is an object of ToyTrain
            self.item = item # <= this is where composition comes in!
            self.item.print_descrption()

    def control_the_train(self):
        self.item.accelerate_train()

    def close_the_box(self):
        if self.is_open:
            self.is_open = False
            del self.item
            print('>> The box is now closed!')
        else:
            print('>> The box is already closed, no need to close again')
```

```python
# instantiate the 'my_magic_box' obj.
my_magic_box = Box()
my_magic_box.open_the_box()
my_magic_box.open_the_box()

# feeding the 'my_train' obj. into 'my_magic_box' obj.
my_train = ToyTrain(color='blue', size='small', max_speed=0.9)
my_magic_box.unpack_item(my_train)

# ... compisition allows us to revoke methods in the composite class
my_magic_box.control_the_train()
# which is equivalent to...
# my_magic_box.item.accelerate_train()

my_magic_box.close_the_box()
my_magic_box.close_the_box()
```

An alternative "has-a" relationship in object-oriented programming, parallel to the concept of composition discussed above, is called **aggregation**. While composition depicts a *strong* association between objects, aggregation represents a *weak* association between the objects. In an aggregation relationship, each class can exist and function independently without relying on one another.
