There exist several ways to declare/initialise a 2-D nested list that represents a rectangular board - **In this course (Programming 2), you are expected to be able to create such a structure using at least one of the three methods mentioned below, and to understand how this method works internally.** ☝

To start, we simply decide the number of rows and number of columns of the board:

```python
N_ROW = 3
N_COL = 4
```

### Method 1: using nested `for` loops

In my opinion, this is the most straightforward strategy to declare the board.

The idea can be broken down into two simple steps:

1. **Create a row**: Start with an empty list and append the same element to it `N_COL` times.
(imagine this is a row vector with $N_{\rm col}$ elements represents $N_{\rm col}$ columns per row.)

2. **Build the board**: Append the initialized row to the main board structure for `N_ROW` times. (so you obtain a $N_{\rm row} \times N_{\rm col}$ matrix.)


```python
board = []
for i in range(N_ROW):
  r = []
  for j in range(N_COL):
    r.append(' . ')
  board.append(r)

print(board)
print('dimension 1:', len(board))
print('dimension 2:', len(board[0]))
```

As you have seen, there are `len(board) = 3`, representing 3 rows (in the first dimension), and `len(board[0]) = 4`, representing 4 columns per row.

So far, it is good enough. If you want to know more, please continue...

### Method 2: using a single `for` loop

This method is very anaglous to the first method. The *only* difference is we replaced the inner `for` loop, which was responsible in creating each row (with `N_COL` elements) of the board, to a simpler expression: `r = [' . ']*N_COL`.

```python
board = []
for i in range(N_ROW):
  r = [' . ']*N_COL
  board.append(r)

print(board)
```

### Method 3: using list comprehension

This method, based on Python's list comprehension, is very commonly seen in the LLM (*e.g.*, ChatGPT) generated solutions.

- `[' . ' for i in range(N_COL)]`: this is the inner list. We used `for` loop creates one row of the board.

- `[ ... for j in range(N_ROW)]`: this is the outer list. We use another `for` loop to repeat the inner list creation process `N_ROW` times.

As I suggested in Lab 2 briefing, albeit the elegance and conciseness in this expression, please use it with *great* discretion, as this compact syntax can obscure the logic and make debugging harder. If you would like to go ahead, make sure you fully understand the idiosyncrasies of the syntax.

```python
board = [[' . ' for i in range(N_COL)] for j in range(N_ROW)]
print(board)
```

One last thing... as you have seen, the dummy indices `i` and `j` accompanied with the `for` loops were not used in list creation. So we may intentionally neglect them, simply by replacing `i` and `j` using underscores `_` - it tells Python to 'dump' this useless variable, and it improves your code readbility!

```python
board = [[' . ' for _ in range(N_COL)] for _ in range(N_ROW)]
print(board)
```
